```python
fruits = ["apple", "banana", "orange", "mango", "watermelon"]
fruits.insert(1, "kiwi")
print(fruits)
fruits.pop()
print(fruits)
fruits.remove("banana")
print(fruits)

print(fruits.count("orange"))

print([len(word) for word in fruits])
```

```python
## List comprehension

lst = []
for x in range(11):
    lst.append(x**2)

print(lst)
```

```python
n = 5
for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1) )
```
