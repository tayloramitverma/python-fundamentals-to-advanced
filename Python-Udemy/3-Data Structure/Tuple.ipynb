```python
## packaging 

tpl_packing = "Amit", "Verma", 1, 2, 3, True, False
print(tpl_packing)
```

```python
start, *middle, end = tpl_packing

print(start)
print(middle)
print(end)
```

```python
lst = [1,2,1,1,1,2,3,4,4,4,4,4,5,5,6,7]
frequency = {}

for number in lst:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
    
print(frequency)
```
