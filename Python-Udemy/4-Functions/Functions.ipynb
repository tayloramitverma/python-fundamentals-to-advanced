```python
def print_details(*args,**kwargs):
    for arg in args:
        print(arg, end= ", ")
    print("\n")
    for k,v in kwargs.items():
        print(f"{k}:{v}")
    

print_details("Amit", "Verma", 1,2,3,4, name="Amit", mob="988775757")
```

```python
## factorial function
def fectorial(n):
    if n==0:
        return 1
    else:
        return n * fectorial(n-1)
    
fectorial(6)
```

```python
def read_count_word(filepath):
    word_count={}
    with open(filepath, 'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = word.lower().strip('.,!?:;"\'')
                word_count[word]=word_count.get(word, 0)+1
    return word_count


read_count_word('simple.txt')
```

```python
## Map in python
lst = ["apple", "banana", "kiwi"]

mapped_lst = list(map(str.upper, lst))

print("Mapped list:", mapped_lst)
```

```python
num_list = [1,2,3,4,5,6]

mapped_num_list = list(map(lambda a:a*a, num_list))
print(mapped_num_list)
```

```python
num_list1 = [1,2,3]
num_list2 = [4,5,6]

mapped_num_list = list(map(lambda x,y:x+y, num_list1, num_list2))
print(mapped_num_list)
```

```python
## Filter in python

lst = list(filter(lambda a:a%2==0, range(11)))
print(lst)

lst = list(filter(lambda a:a%2==0 and a>4, range(11)))
print(lst)
```
