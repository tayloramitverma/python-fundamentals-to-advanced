### Polymorphism


```python
from abc import ABC, abstractmethod

class Vahicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vahicle):
    def start_engine(self):
        return "Car engine started!"
    
class MoterCycle(Vahicle):
    def start_engine(self):
        return "MoterCycle engine started!"
    
# Function that demonstrates polymorephism

def start_engine(vehicle):
    print(vehicle.start_engine())


car = Car()
moter_cycle = MoterCycle()

start_engine(car)
start_engine(moter_cycle)
```
