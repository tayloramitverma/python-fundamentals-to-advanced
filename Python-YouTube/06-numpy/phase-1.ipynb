## List Vs Array


```python
import time

# List
start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n Python List operation time: ", time.time() - start)


# Numpy Array
start = time.time()
np_arr = np.arange(1000000) * 2
print("\n Numpy List operation time: ", time.time() - start)
```

```python
print("zeros array: \n", np.zeros((3,4)))
print("ones array: \n", np.ones((3,4)))
print("full array: \n", np.full((3,4), 5))
print("random array: \n", np.random.random((2,3)))
print("sequence array: \n", np.arange(0, 11, 1))
```

### Array Reshaping

```python
np_arr = np.arange(12)

print("Original Numpy array - (Vector): ", np_arr)

reshaped = np_arr.reshape((3,4))
print("\n Reshaped array:", reshaped)

flattened = reshaped.flatten()
print("\n Flatten array:", flattened)

# ravel (Retun view instead of copy)
raveled = reshaped.ravel()
print("\n Ravel array:", raveled)

transposed = reshaped.T
print("\n Transpose array:", transposed)
```
