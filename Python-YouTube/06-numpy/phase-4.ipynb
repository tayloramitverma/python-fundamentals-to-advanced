```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
array1 = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
array2 = np.random.rand(3, 3)
array3 = np.zeros((4, 4))

## Save in file
np.save("array1.npy", array1)
np.save("array2.npy", array2)
np.save("array3.npy", array3)


## load from file

arr1 = np.load("array1.npy")

print(arr1)
```

```python
try:
    logo = np.load("array1.npy")

    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Array1 Logo")
    plt.grid(False)

    dark_logo = 1 - logo

    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Array1 Dark Logo")
    plt.grid(False)


except FileNotFoundError:
    print("Logo not found")
```

```python
img = Image.open("NumPy.png")
print(type(img))
print(img)

# summarize some details about the image
print(img.format)
print(img.size)
print(img.mode)
```

```python
img_np_array = np.array(img)
print(type(img_np_array))
print(img_np_array.shape)
```

```python
# convert the image to RGBA first, then drop the alpha channel
img_rgba = img.convert("RGBA")
img_np_array = np.array(img_rgba)

rgb_array = np.delete(arr=img_np_array, obj=3, axis=2)
print(rgb_array.shape)
```

```python
plt.imshow(rgb_array)
plt.show()

dark_logo = 1 - rgb_array

plt.imshow(dark_logo)
plt.show()
```
